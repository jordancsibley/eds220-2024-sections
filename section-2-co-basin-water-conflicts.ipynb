## String accessor for `pandas.Series`

## Wrangling

```python
string = df['State'].str.split(';', expand=True).stack().str.strip
```

```python
# select state column 
df["State"]
```

```python
# spilt the strings by the ;
# expand=True makes it a df where the spilts get there own column
df["State"].str.split(";", expand=True)
```

```python
# stack all the observations into a list 
df["State"].str.split(";", expand=True).stack()
```

```python
# gives the unique values, see that there are some repeats due to the white spaces
df["State"].str.split(";", expand=True).stack().unique()
```

```python
# remove the white space using str.strip()
df["State"].str.split(";", expand=True).stack().str.strip().unique()
```

The default of `str.strip()` is to remove the white spaces, but you can set it to be what you want 

Now how would you see how many times the states show up?

```python
df["State"].str.split(";", expand=True).stack().str.strip().value_counts()
```
